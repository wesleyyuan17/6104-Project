```python
def evaluate(agent, val_data, n_steps=None, plot=False, save_plots=False, env_type='no_restrictions'):
    # Test the trained agent
    val_env = SingleStockTradingEnv('../data/spy_dev.csv', 
                                    initial_value=initial_port_value, 
                                    borrowing=borrowing, 
                                    long_only=long_only)
    obs = val_env.reset()
    if n_steps is None:
        n_steps = val_data.shape[0] - 6
    portfolio_values = []
    agent_holdings = []
    total_reward = 0
    for step in range(n_steps):
        action, _ = agent.predict(obs, deterministic=True)
        # print("Step {}".format(step + 1))
        # print("Action: ", action)
        obs, reward, done, info = val_env.step(action)
        if not np.isnan(reward):
            portfolio_values.append(val_env.agent_portfolio_value)
            total_reward += reward
            agent_holdings.append(val_env.agent_shares)
        # print('reward=', reward, 'done=', done)
        # env.render(mode='console')
        if done:
            # Note that the VecEnv resets automatically
            # when a done signal is encountered
            print("Goal reached!", "reward=", reward)
            break

    print(total_reward)

    # print returns over time period
    print(portfolio_values[-1] / initial_port_value, \
          # agent_portfolio_values[1][-1] / initial_port_value, \
          np.exp(val_data['Returns'][:n_steps].cumsum()).iloc[-1])
    
    if plot:
        # visualize PnL
        fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20,10))
        axes[0].plot(val_data['Date'][:n_steps], np.exp(val_data['Returns'][:n_steps].cumsum())*initial_port_value, label='Buy and Hold')
        axes[0].plot(val_data['Date'][:n_steps], portfolio_values, label='A2C')
        axes[0].legend()
        
        axes[1].plot(val_data['Date'][:n_steps], agent_holdings, label='A2C', marker='o')
        if save_plots:
            fig.savefig('figures/rl_performance_comparison_{}.jpg'.format(env_type), bbox_inches='tight')
        plt.show()
```
