```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn

from stable_baselines3 import A2C

import gpytorch

from custom_env import *
from model import *
from data_util import *
```

```python
# plt.style.available
```

```python
initial_port_value = 1000
borrowing = True
long_only = False
```

```python
train_data = pd.read_csv('../data/spy_train.csv')
val_data = pd.read_csv('../data/spy_validation.csv')
val_data['Date'] = pd.to_datetime(val_data['Date'])

env = SingleStockTradingEnv('../data/spy_train.csv', 
                            engineer_features,
                            initial_value=initial_port_value, 
                            borrowing=borrowing,
                            long_only=long_only)
```

```python
a2c_agent = A2C.load('checkpoints/a2c_no_restrictions_v2') # A2C('MlpPolicy', env, gamma=0.1)
rl_portfolio_values, rl_agent_holdings, rl_agent_actions, rl_goal_num_shares, rl_fig = evaluate(a2c_agent, 
                                                                                                'SPY',
                                                                                                val_data, 
                                                                                                initial_port_value, 
                                                                                                borrowing, 
                                                                                                long_only, 
                                                                                                use_gp=False, 
                                                                                                plot=True, 
                                                                                                save_plots=False, 
                                                                                                env_type='no_restrictions')
```

```python
lookback = 5
train_features = engineer_features(train_data, lookback=lookback)
# train_features['shares'] = 0 # to match dimensions for GP later on
# train_features['cash'] = 0 # to match dimensions for GP later on
train_features.head()
```

```python
# turn data in dataframes into model inputs
X_train = torch.Tensor(train_features.drop(['Date', 'Volume', 'Returns', 'Close', f'Open -{lookback}'], axis=1).values)
y_train = torch.Tensor(train_features['Returns'].values)

gp_params = {'n_train': 20,
             'training_iter': 10,
             'cvar_limit': -5, # maximum loss tolerance %
             'gp_limit': 0.3, # predicted magnitude of GPR such that GP takes over
             'data': {'X_train': X_train[-20:], 'y_train': y_train[-20:]} # last month's worth of data?
             }
a2c_gp_agent = TradingAgent(use_gp=True, gp_params=gp_params, policy='MlpPolicy', env=env)
a2c_gp_agent.load(rl_path='checkpoints/a2c_no_restrictions_v2')
```

```python
a2c_gp_agent.learn(5000)
```

```python
gp_portfolio_values, gp_agent_holdings, gp_agent_actions, gp_goal_num_shares, gp_fig = evaluate(a2c_gp_agent, 
                                                                                                'SPY',
                                                                                                val_data,  
                                                                                                initial_port_value, 
                                                                                                borrowing, 
                                                                                                long_only, 
                                                                                                use_gp=True, 
                                                                                                plot=True, 
                                                                                                save_plots=False, 
                                                                                                env_type='no_restrictions')
```

```python
comp_fig = plt.figure(figsize=(20,5))
plt.plot(val_data['Date'].iloc[6:], np.exp(val_data['Returns'].iloc[6:].cumsum())*initial_port_value, label='Buy and Hold')
plt.plot(val_data['Date'].iloc[6:], rl_portfolio_values, label='A2C')
plt.plot(val_data['Date'].iloc[6:], gp_portfolio_values, label='A2C + GP')
plt.title('Performance Comparison - SPY')
plt.xlabel('Date')
plt.ylabel('Portfolio Value')
plt.legend()
plt.show()
```

```python
rl_fig.savefig('figures/rl_base_no_restrictions.pdf', bbox_inches='tight')
gp_fig.savefig('figures/rl_with_gp_no_restrictions.pdf', bbox_inches='tight')
comp_fig.savefig('figures/rl_gp_comparison_no_restrictions.pdf', bbox_inches='tight')
```

```python
actions_fig = plt.figure(figsize=(20,5))
plt.bar(val_data['Date'].iloc[6:], rl_agent_actions, label='RL Actions', alpha=0.75)
plt.bar(val_data['Date'].iloc[6:], gp_agent_actions, label='GP Actions', alpha=0.75)
plt.title('Actions Comparison - SPY')
plt.legend()
plt.show()

actions_fig.savefig('figures/actions_comparison_no_restrictions.pdf', bbox_inches='tight')
```

```python
shares_fig = plt.figure(figsize=(20,5))
# plt.plot(range(len(rl_goal_num_shares)), rl_goal_num_shares, label='RL Target # Shares')
plt.fill_between(val_data['Date'].iloc[6:], rl_agent_holdings, label='RL Current # Shares', alpha=0.5)
# plt.bar(val_data['Date'].iloc[6:], gp_goal_num_shares, label='GP Target # Shares', alpha=0.75)
plt.fill_between(val_data['Date'].iloc[6:], gp_agent_holdings, label='GP Current # Shares', alpha=0.5)
plt.title('Holdings Comparison - SPY')
plt.legend(loc='lower right')
plt.show()

shares_fig.savefig('figures/num_shares_comparison_no_restrictions.pdf', bbox_inches='tight')
```

```python
results = pd.DataFrame({'Date': val_data['Date'][6:], 
                        'Buy and Hold': val_data['Adj Close'][6:].values, 
                        'A2C': rl_portfolio_values, 
                        'A2C + GP': gp_portfolio_values})
results.reset_index(drop=True, inplace=True)
for col in results.drop('Date', axis=1).columns:
    roll_max = results[col].rolling(252, min_periods=1).max()
    daily_drawdown = results[col]/roll_max - 1.0
    results[col + ' Drawdown'] = daily_drawdown
results.head()
```

```python
results.min()
```

```python
fig = results.set_index('Date')[['Buy and Hold Drawdown', 'A2C Drawdown', 'A2C + GP Drawdown']].plot(figsize=(20,5), rot=0)
plt.show()
```

```python
results.to_csv('../data/spy_results.csv', index=False)
```
