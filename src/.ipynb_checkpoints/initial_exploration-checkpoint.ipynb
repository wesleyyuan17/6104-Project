```python
# simplest simulation - assume returns are student-t, volatility is 10-day historical moving average
# doesn't take advantage of all available info in historical data
def sample_returns(N, hist_vol, n_lb=10, lam=0.1):
    sample = [0]
    lookback = []
    t_dist = t(n_lb-1, 0, hist_vol)
    for i in range(n_lb):
        lookback.append(t_dist.rvs())
    for i in range(N):
        temp_vol = np.std(lookback) - lam*(np.std(lookback) - hist_vol) # added mean-reverting
        r = t.rvs(n_lb-1, 0, temp_vol)
        lookback.pop(0)
        lookback.append(r)
        sample.append(r)
    return sample[1:]
```

```python
n_lb = 20 # 1 month look back
lam = 0.05
train['Returns'].plot(figsize=(20,10))
sim_returns_sample = []
for i in range(10):
    sim_returns = sample_returns(dev.shape[0], np.std(train['Returns']), n_lb, lam)
    plt.plot(dev.index, sim_returns, alpha=0.5)
    sim_returns_sample.append(sim_returns)
plt.show()
```

```python
print(np.std(train['Returns']), np.mean([np.std(sr) for sr in sim_returns_sample]))
print([np.std(sr) for sr in sim_returns_sample])
```
